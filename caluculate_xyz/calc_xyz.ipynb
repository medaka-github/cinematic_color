```python
light_skin = colour.SDS_COLOURCHECKERS['ColorChecker N Ohta']['light skin']
```

```python
ilm = colour.SDS_ILLUMINANTS['D65']
```

```python
cmfs = colour.MSDS_CMFS.get('CIE 1931 2 Degree Standard Observer')
```

```python
shape = colour.SpectralShape(380, 780, 10)
light_skin = light_skin.align(shape)
light_skin = light_skin.align(shape)
ilm = ilm.align(shape)
cmfs = cmfs.align(shape)
```

```python
colour.sd_to_XYZ(light_skin, cmfs, ilm)/100
```

```python
colour.sd_to_XYZ(light_skin, cmfs)
```

```python
colour.sd_to_XYZ(light_skin, cmfs, method='ASTM E308', shape = shape)
```

```python
sd = light_skin * ilm
```

```python
colour.sd_to_XYZ(sd, cmfs, method='Integration', shape = shape)
```

```python
colour.sd_to_XYZ(sd, cmfs, method='ASTM E308', shape = shape)
```

```python
colour.sd_to_XYZ(light_skin, cmfs, ilm, method='Integration', shape = shape)
```

```python
colour.colorimetry.sd_to_XYZ_integration(sd)
```
